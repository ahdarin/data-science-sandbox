```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Menambahkan dataset

```python
df = pd.read_csv('/content/Online Sales Data.csv')
```

Menampilkan lima baris pertama dari dataset

```python
df.head()
```

Menghitung statistik deskriptif untuk setiap fitur

```python
df.describe()
```

Memeriksa Missing Value

```python
df.isnull().sum()
```

Analisis untuk kolom-kolom numerik dengan menampilkan histogram dan boxplot.

```python
numerical_cols = ['Units Sold', 'Unit Price', 'Total Revenue']
df.hist(bins=20, figsize=(10, 8))
plt.show()
```

```python
df.plot(kind='box', subplots=True, layout=(2,2), figsize=(10,8))
plt.show()
```

Untuk data kategorikal, hitung frekuensi nilai dan tampilkan dalam bentuk bar plot.

```python
print(df['Product Category'].value_counts())
```

```python
df['Product Category'].value_counts().plot(kind='bar')
plt.show()
```

Memeriksa apakah ada outliers di kolom-kolom numerik menggunakan boxplot.

```python
sns.boxplot(data=df[numerical_cols])
plt.show()
```

Visualisasi Matriks korelasi untuk kolom-kolom numerik menggunakan heatmap.

```python
corr_matrix = df[numerical_cols].corr()
print(corr_matrix)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```

Membuat grafik untuk memvisualisasikan distribusi salah satu fitur. Buat scatter plot untuk melihat hubungan antara dua variabel numerik.

```python
sns.histplot(df[numerical_cols[2]], kde=True)
plt.show()
```

```python
plt.scatter(df[numerical_cols[1]], df[numerical_cols[2]])
plt.xlabel(numerical_cols[1])
plt.ylabel(numerical_cols[2])
plt.title(f"Scatter plot of {numerical_cols[1]} vs {numerical_cols[2]}")
plt.show()
```
