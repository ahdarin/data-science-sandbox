```python
import pandas as pd #untuk membaca dataset
from sklearn.model_selection import train_test_split #untuk memisahkan data training dan data test
from sklearn.tree import DecisionTreeClassifier #library model decision tree
from sklearn.preprocessing import LabelEncoder #untuk mengubah String ke Int
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix #untuk menghitung akurasi
```

```python
dataset = pd.read_csv('/content/nilai-utbk.csv')
```

```python
dataset.head()
```

```python
label_encoder = LabelEncoder()
dataset['pilihan1_encoded'] = label_encoder.fit_transform(dataset['PILIHAN 1'])
dataset['pilihan2_encoded'] = label_encoder.fit_transform(dataset['PILIHAN 2'])
```

```python
dataset.head()
```

```python
X=dataset.drop(['KELULUSAN','PILIHAN 1','PILIHAN 2'],axis=1)
Y=dataset.KELULUSAN
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1)
```

```python
dt=DecisionTreeClassifier() #objek untuk decisiontree
```

```python
dt.fit(X_train,Y_train) #train model decision tree classifier
```

```python
Y_pred=dt.predict(X_test) #prediksi hasil data test
```

```python
print('Akurasi:',accuracy_score(Y_test,Y_pred))
```

```python
print('Hasil:',classification_report(Y_test,Y_pred))
```
